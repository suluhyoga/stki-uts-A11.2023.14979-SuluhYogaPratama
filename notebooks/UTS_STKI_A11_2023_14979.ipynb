## Boolean Retrieval Model

```python
from src.boolean_ir import load_processed_docs, build_inverted_index, boolean_search

docs = load_processed_docs('data/processed')
index = build_inverted_index(docs)
all_docs = list(docs.keys())

# uji beberapa query
print("Query 1:", boolean_search("informasi AND sistem", index, all_docs))
print("Query 2:", boolean_search("NOT evaluasi", index, all_docs))
print("Query 3:", boolean_search("(indeks AND vektor) OR evaluasi", index, all_docs))
```

## Vector Space Model (TF-IDF)

```python
from src.vsm_ir import compute_tfidf, search_vsm

tfidf_docs, idf = compute_tfidf(docs)
hasil = search_vsm("informasi sistem dokumen", tfidf_docs, idf)
hasil[:10]
```

## Evaluasi (Precision, Recall, MAP, nDCG)

```python
from src.eval import precision_recall, precision_at_k, average_precision, ndcg_at_k

gold = ['doc1.txt', 'doc2.txt', 'doc3.txt']
res = [h[0] for h in hasil[:5]]

p, r, tp, fp, fn = precision_recall(res, gold)
print("Precision:", p, "Recall:", r)
print("P@5:", precision_at_k(hasil, gold, k=5))
print("AP:", average_precision(hasil, gold))
print("nDCG:", ndcg_at_k(hasil, gold))
```

## Penjelasan Hasil Evaluasi

Berdasarkan hasil evaluasi menggunakan metrik Information Retrieval, sistem memperoleh Precision sebesar 0.20 dan Recall sebesar 0.33. Nilai-nilai tersebut menunjukkan bahwa sistem masih cenderung menampilkan dokumen yang tidak relevan serta belum mampu menemukan seluruh dokumen relevan yang ada. Nilai AP (Average Precision) sebesar 0.33 dan nDCG (Normalized Discounted Cumulative Gain) sebesar 0.47 mengindikasikan bahwa urutan hasil pencarian masih perlu ditingkatkan agar dokumen relevan dapat muncul pada peringkat lebih tinggi.
